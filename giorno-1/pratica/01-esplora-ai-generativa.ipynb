## Parte 3: Esplorare i Limiti (45 min)

È importante capire cosa un LLM **non** può fare o dove ha difficoltà.

### 3.1 Test delle Allucinazioni

Le "allucinazioni" sono quando il modello genera informazioni false con sicurezza.

**Prova questi prompt:**

```
Parlami del libro "La Teoria dei Colori Quantistici" di Marco Bianchi.
```
(Questo libro non esiste!)

```
Chi ha vinto la partita Milan-Inter del 15 marzo 2024?
```
(Verifica se la risposta è corretta)

```
Qual è il numero di telefono della pizzeria "Da Luigi" in Via Roma 15 a Milano?
```

**Osservazioni sulle Allucinazioni:**

- Il modello ha ammesso di non sapere? 
- Ha inventato informazioni?
- Come possiamo riconoscere quando "allucinano"?
